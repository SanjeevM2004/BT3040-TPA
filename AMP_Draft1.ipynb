```python
import pandas as pd
new_file_path = 'general_amps.xlsx'
new_data = pd.read_excel(new_file_path)

new_data_info = new_data.info()
new_data_head = new_data.head()

new_data_info, new_data_head
```

```python
def not_amp(s):
    if 'Antimicrobial' not in s:
        print(s)
new_data['Activity'].apply(not_amp)
```

```python
new_data['Is_Antimicrobial'] = new_data['Activity'].apply(lambda x: 'Antimicrobial' in x)

new_target_distribution = new_data['Is_Antimicrobial'].value_counts(normalize=True)
new_target_distribution
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def kmerize(sequence, size=3):
    return [sequence[i:i+size] for i in range(len(sequence) - size + 1)]

new_data['Sequence_kmer'] = new_data['Sequence'].dropna().apply(lambda x: ' '.join(kmerize(x)))

vectorizer = CountVectorizer()
X_kmer = vectorizer.fit_transform(new_data['Sequence_kmer'].dropna())

source_encoded = pd.get_dummies(new_data['Source'], prefix='Source')

features = pd.concat([source_encoded, new_data[['Sequence_Length']]], axis=1)

features = features.loc[new_data['Sequence_kmer'].dropna().index]

y = new_data['Is_Antimicrobial'][new_data['Sequence_kmer'].dropna().index]

X_kmer.shape, features.shape, y.shape
```
