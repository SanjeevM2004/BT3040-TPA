```python
import pandas as pd
import propy.CTD as pc
import propy.PseudoAAC as pp
import propy.AAComposition as pa
import protpy as pt
from concurrent.futures import ThreadPoolExecutor

# List of standard amino acids
standard_amino_acids = set('ACDEFGHIKLMNPQRSTVWY')

# Define the structure of the zero-filled feature dictionary based on actual feature names
def get_zero_feature():
    zero_feature = {}
    zero_feature.update(pp.GetAPseudoAAC("ACDEFGHIKLMNPQRSTVWY", lamda=4, weight=0.05))
    zero_feature.update(pp.GetAAComposition("ACDEFGHIKLMNPQRSTVWY"))
    zero_feature.update(pc.CalculateC("ACDEFGHIKLMNPQRSTVWY"))
    # zero_feature.update(pa.GetSpectrumDict("ACDEFGHIKLMNPQRSTVWY"))  # Uncomment if using this
    for key in zero_feature:
        zero_feature[key] = 0
    return zero_feature

zero_feature = get_zero_feature()

# Define the descriptor_calc function
def descriptor_calc(row):
    sequence = row['Sequence']
    label = row['Labels']
    feature = {}
    if not set(sequence).issubset(standard_amino_acids):
        feature = zero_feature.copy()  # Ensure a unique copy for each row
        feature['Labels'] = label
        return feature
    if len(sequence) > 4:
        feature.update(pp.GetAPseudoAAC(sequence, lamda=4, weight=0.05))
        feature.update(pp.GetAAComposition(sequence))
        feature.update(pc.CalculateC(sequence))
        # feature.update(pa.GetSpectrumDict(sequence))  # Uncomment if using this
        feature['Labels'] = label
    return feature

# Function to apply descriptor_calc in parallel
def process_row(row):
    return descriptor_calc(row)

# Apply the descriptor_calc function to each row using multithreading
with ThreadPoolExecutor(max_workers=10) as executor:  # Adjust max_workers as needed
    features_list = list(executor.map(process_row, [row for _, row in data.iterrows()]))

# Convert the resulting list of dictionaries to a DataFrame
features_df = pd.DataFrame(features_list)

# Display the new DataFrame
features_df

```

```python
features_df.to_excel('data/Dataset_extracted.xlsx')
```

```python
train_data = pd.read_excel('data/AMP_BPM_train_data.xlsx')
train_data = train_data.drop('Unnamed: 0', axis=1)

# Define the descriptor_calc function
def descriptor_calc(row):
    sequence = row['Sequence']
    label = row['Label']
    feature = {}
    feature.update(pp.GetAPseudoAAC(sequence, lamda=4, weight=0.05))
    feature.update(pp.GetAAComposition(sequence))
    #feature.update(pt.pseudo_amino_acid_composition(sequence, lamda=4))
    feature.update(pc.CalculateC(sequence))
    #feature.update(pa.GetSpectrumDict(sequence))
    feature['Label'] = label
    return feature

# Apply the descriptor_calc function to each row
features_train_df = train_data.apply(descriptor_calc, axis=1)

# Convert the resulting Series of dictionaries to a DataFrame
features_train_df = pd.DataFrame(features_train_df.tolist())

features_train_df.to_excel('data/train_df_bpm.xlsx')
```
