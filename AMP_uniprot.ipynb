```python
data = pd.read_excel('data/Dataset.xlsx')
```

```python
df_concatenated['Sequence'] = data['Sequence']
```

```python
df_concatenated = df_concatenated.drop('Unnamed: 0', axis=1)
```

```python
# Reorder columns to move 'Sequence' to the first position
cols = ['Sequence'] + [col for col in df_concatenated.columns if col != 'Sequence']
df_concatenated = df_concatenated[cols]
```

```python
df_concatenated
```

```python

# Shuffle the rows
df_shuffled = df_concatenated.sample(frac=1.0, random_state=42)  # Set random_state for reproducibility

# Reset the index
df_shuffled.reset_index(drop=True, inplace=True)

# Display the shuffled dataframe
df_shuffled
```
