```python
from Bio import SeqIO
import pandas as pd

# Function to read FASTA file and extract sequences
def read_fasta(file_path, num_sequences):
    sequences = []
    with open(file_path, "r") as fasta_file:
        for record in SeqIO.parse(fasta_file, "fasta"):
            sequences.append(str(record.seq))
            if len(sequences) == num_sequences:
                break
    return sequences

# Path to the FASTA file
fasta_file_path = "uniprotkb_length_5_TO_100_AND_keyword_K_2024_04_30.fasta"  # Replace with the path to your FASTA file

# Number of sequences to extract
num_sequences = 9000

# Read FASTA file and extract sequences
sequences = read_fasta(fasta_file_path, num_sequences)

# Create DataFrame
df = pd.DataFrame({"Sequence": sequences, "Label": "Antimicrobial"})

# Display DataFrame
print(df.head())
```

```python
from Bio import SeqIO
import pandas as pd

# Function to read FASTA file and extract sequences
def read_fasta(file_path, num_sequences):
    sequences = []
    with open(file_path, "r") as fasta_file:
        for record in SeqIO.parse(fasta_file, "fasta"):
            sequences.append(str(record.seq))
            if len(sequences) == num_sequences:
                break
    return sequences

# Path to the FASTA file
fasta_file_path = "uniprotkb_length_70_TO_80_NOT_keyword_K_2024_04_30.fasta"  # Replace with the path to your FASTA file

# Number of sequences to extract
num_sequences = 9000

# Read FASTA file and extract sequences
sequences = read_fasta(fasta_file_path, num_sequences)

# Create DataFrame
df_am = pd.DataFrame({"Sequence": sequences, "Label": "Non-Antimicrobial"})

# Display DataFrame
print(df_am.head())
```

```python
df_concatenated = pd.concat([df, df_am], axis=0)
```

```python
import pandas as pd

# Assuming df is your dataframe
# Shuffle the rows
df_shuffled = df_concatenated.sample(frac=1.0, random_state=42)  # Set random_state for reproducibility

# Reset the index
df_shuffled.reset_index(drop=True, inplace=True)

# Display the shuffled dataframe
print(df_shuffled)
```
