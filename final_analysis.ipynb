```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from catboost import CatBoostClassifier
import joblib

data = data.dropna()
X = data.drop(columns=['Labels', 'Sequence'])
y = data['Labels']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8)

# Initialize models
svm_classifier = SVC(probability=True)
logistic_regression = LogisticRegression(max_iter=1000)
decision_tree = DecisionTreeClassifier()
catboost_classifier = CatBoostClassifier(iterations=300, depth=10, learning_rate=0.1, verbose=0)

# Train models
svm_classifier.fit(X_train, y_train)
logistic_regression.fit(X_train, y_train)
decision_tree.fit(X_train, y_train)
catboost_classifier.fit(X_train, y_train)

    # Predictions (optional)
dt_pred = decision_tree.predict(X_test)
svm_predictions = svm_classifier.predict(X_test)
logistic_regression_predictions = logistic_regression.predict(X_test)
catboost_predictions = catboost_classifier.predict(X_test)

svm_accuracy = accuracy_score(y_test, svm_predictions)
print("SVM Accuracy:", svm_accuracy)
# Calculate accuracy for Logistic Regression
logistic_regression_accuracy = accuracy_score(y_test, logistic_regression_predictions)
print("Logistic Regression Accuracy:", logistic_regression_accuracy)

# Evaluate the accuracy of the model
dt_accuracy = accuracy_score(y_test, dt_pred)
print("Decision tree Accuracy:", dt_accuracy)

# Calculate accuracy
accuracy = accuracy_score(y_test, catboost_predictions)
print("Catboost Accuracy:", accuracy)

# Save models
print('Saving Models')
joblib.dump(svm_classifier, 'svm_model.pkl')
joblib.dump(logistic_regression, 'logistic_regression_model.pkl')
joblib.dump(decision_tree, 'decision_tree_model.pkl')
joblib.dump(catboost_classifier, 'catboost_model.pkl')

# Apply PCA and train models on PCA-transformed data
for pca_value in [5, 10, 20]:
    pca = PCA(n_components=pca_value)
    x_train_pca = pca.fit_transform(X_train)
    x_test_pca = pca.transform(X_test)
    
    # Train models with PCA
    decision_tree.fit(x_train_pca, y_train)
    svm_classifier.fit(x_train_pca, y_train)
    logistic_regression.fit(x_train_pca, y_train)
    
    # Predictions (optional)
    dt_pred = decision_tree.predict(x_test_pca)
    svm_predictions = svm_classifier.predict(x_test_pca)
    logistic_regression_predictions = logistic_regression.predict(x_test_pca)
    
    print(f"Number of PCA components : {pca_value}")
    # Calculate accuracy for SVM
    svm_accuracy = accuracy_score(y_test, svm_predictions)
    print("SVM Accuracy:", svm_accuracy)

    # Calculate accuracy for Logistic Regression
    logistic_regression_accuracy = accuracy_score(y_test, logistic_regression_predictions)
    print("Logistic Regression Accuracy:", logistic_regression_accuracy)
    
    # Evaluate the accuracy of the model
    dt_accuracy = accuracy_score(y_test, dt_pred)
    print("Decision tree Accuracy:", dt_accuracy)
    # Save PCA models
    print('Saving Models')
    joblib.dump(pca, f'pca_{pca_value}.pkl')
    joblib.dump(decision_tree, f'decision_tree_pca_{pca_value}.pkl')
    joblib.dump(svm_classifier, f'svm_pca_{pca_value}.pkl')
    joblib.dump(logistic_regression, f'logistic_regression_pca_{pca_value}.pkl')

```
